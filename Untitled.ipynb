```python
cur.execute("""
            select * from Matches as M
            join Teams_in_Matches as TM on M.Match_ID = TM.Match_ID
            join Unique_Teams as UT on TM.Unique_Team_ID = UT.Unique_Team_ID
            where Date between '2011-08-01' and '2013-05-31'
            order by TM.Match_ID asc
            """).fetchmany(1)
```

```python
df_M = pd.DataFrame(cur.execute(""" select * from Matches where Date between '2011-08-01' and '2013-05-31' order by Date asc """).fetchall())
df_M.columns = ['Match_ID', 'Div', 'Season', 'Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR']

df_UT = pd.DataFrame(cur.execute(""" select * from Unique_Teams """).fetchall())
df_UT.columns = ['TeamName', 'Unique_Team_ID']

df_TM = pd.DataFrame(cur.execute("""select * from Teams_in_Matches """).fetchall())
df_TM.columns = ['Match_ID', 'Unique_Team_ID']


df_M['home_count'] = df_M['FTR'] == 'H'
df_M['away_count'] = df_M['FTR'] == 'A'
# df_M.head(10)
df_h = df_M.groupby('HomeTeam').agg({'home_count':'sum'})
# df_h

df_a = df_M.groupby('AwayTeam').agg({'away_count': 'sum'})
df_a
# print(df_UT.head(1))
# print(df_TM.head(1))
```
